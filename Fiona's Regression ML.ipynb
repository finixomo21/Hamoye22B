```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
energy= pd.read_csv('energydata_complete.csv')
energy.head
```

```python
energy.columns
```

```python
column_names = {'T1':'temp_kitchen', 'RH_1':'humidity_kitchen', 'T2':'temp_livingrm', 'RH_2':'humidity_livingrm', 'T3':'temp_laundryrm', 'RH_3':'humidity_laundryrm', 'T4':'temp_office', 'RH_4':'humidity_office', 'T5':'temp_bathrm', 'RH_5':'humidity_bathrm', 'T6':'temp_outside', 'RH_6':'humidity_outside', 'T7':'temp_ironingrm', 'RH_7':'humidity_ironingrm', 'T8':'temp_teenagerm',
       'RH_8':'humidity_teenagerm', 'T9':'temp_parentsrm', 'RH_9':'humidity_parentsrm', 'T_out':'temp_outcherves','RH_out':'humidity_outcherves'}
```

```python
energy.rename(columns=column_names, inplace=True)
energy
```

```python
energy.describe()
```

```python
#select a sample of the dataset
simple_linear_reg_df = energy[['temp_livingrm', 'temp_outside']].sample(15, random_state=42)
simple_linear_reg_df
```

```python
energy2=energy.drop(['date','lights'],axis=1)
energy2
```

```python
scaler = MinMaxScaler()
normalised_energy = pd.DataFrame(scaler.fit_transform(energy2), columns=energy2.columns)
features_energy = normalised_energy.drop(columns=['Appliances'])
heating_target = normalised_energy['Appliances']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_energy, heating_target,
test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
#obtain predictions
predicted_values = linear_model.predict(x_test)
#MAE
mae = mean_absolute_error(y_test, predicted_values)
round(mae,2) 
```

```python
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
#this function returns the weight of every feature
 weights = pd.Series(model.coef_, feat.columns).sort_values()
 weights_df = pd.DataFrame(weights).reset_index()
 weights_df.columns = ['Features', col_name]
 weights_df[col_name].round(3)
 return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
